```python
#EXPLORATORY DATA ANALYSIS
#GENDER
df.columns
```

```python
ax = sns.countplot(x='Gender',data=df) #to plot of the purchase of genders

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x='Gender',y='Amount',data=sales_gen)
```

```python
#Analysis on age
df.columns
```

```python
ax = sns.countplot(x='Age Group',hue='Gender',data=df) #graphical repesentation of specific age groups as well as Genders who purchased the items 

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Total amount vs age group
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x='Age Group',y='Amount',data=sales_age)
```

```python
#State wise data analysis
df.columns
```

```python
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False)#it shows number of orders statewise

sns.set(rc={'figure.figsize':(30,20)})
sns.barplot(x='State',y='Orders',data=sales_state)

```

```python
#Total amount spend statewise
sales_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.set(rc={'figure.figsize':(30,20)})
sns.barplot(x='State',y='Amount',data=sales_state)
```

```python
#Marital Status
ax = sns.countplot(x='Marital_Status',data=df)#on the basis of marital status

sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_Mar=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.set(rc={'figure.figsize':(7,5)})
sns.barplot(x='Marital_Status',y='Amount',hue='Gender',data=sales_Mar)
```

```python
#OCCUPATION
ax = sns.countplot(x='Occupation',data=df)#on basis of occupation who ordered most items

sns.set(rc={'figure.figsize':(30,20)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#PRODUCT CATEGORY
ax = sns.countplot(x='Product_Category',data=df)#on the basis of category of product

sns.set(rc={'figure.figsize':(30,20)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_Prod=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.set(rc={'figure.figsize':(40,20)})
sns.barplot(x='Product_Category',y='Amount',data=sales_Prod)
```

```python
#top 10 most sold products along with Product ID
fig1,ax1 = plt.subplots(figsize=(12,7))
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')

```
